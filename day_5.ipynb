```python
def balancedStringSplit(s):
    c,res=0,0
    for i in s:
        if i=="R":
            c+=1
        else:
            c-=1
        if c==0:
            res+=1
    return res
```

```python
class Solution:
    def balancedStringSplit(self, s: str) -> int:
        c=0
        lst=[]
        a=""
        for i in s:
            a+=i
            if len(a)%2==0 and a.count("L") ==a.count("R"):
                lst.append(a)
            else:
                continue
        return len(lst)
```

```python
def numberOfSpecialChars(word):
    c=0
    lst=[]
    for i in word:
        if i.islower() and i not  in lst:
            if i.upper() and i.lower() in lst:
                c-=1
            else:
                if i.upper() in lst:
                    continue
                else:
                    lst.append(i)
                    lst.append(i.upper())
                    c+=1
        else:
            lst.append(i)
    return abs(c)

word="AbBCab"   
print(numberOfSpecialChars(word))
```

```python
def cou(word):
    c=set()
    l=set()
    for i in word:
        if i.islower():
            l.add(i)
        elif i.isupper() and i.lower() in l:
            c.add(i.lower())
    return len(c)
word="AbBcCa"   
print(cou(word))
```
